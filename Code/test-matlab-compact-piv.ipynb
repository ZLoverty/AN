# Test Matlab compact PIV

This notebook tests the Matlab implementation of compact PIV. The idea is to generate .mat files using both Python and Matlab methods from the same PIV data, and then compare the quiver plots.

## 1 Python

We need to convert the "_1.txt" file format to my conventional PIV format first. Then we read the data as filelist and construct compact PIV.

```python
from myImageLib import readdata
from pivLib import compact_PIV
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
l = readdata("test_files/pivtxtpy/")
cpiv = compact_PIV(l)
```

```python
cpiv.to_mat("test_files/pivtxtpy.mat")
```

```python
cpiv.get_labels()
```

```python
x, y, u, v = cpiv.get_frame(4)
```

```python
plt.quiver(x, y, u, v)
```

## 2 Matlab

We load the .mat file generated by Matlab code and then plot quiver.

```python
cpiv_dict = loadmat("test_files/pivtxt.mat")
cpivmat = compact_PIV(cpiv_dict)
```

```python
x, y, u, v = cpivmat.get_frame(4)
```

```python
plt.quiver(x, y, u, v)
```

Compare the two quiver plots and we know that the Matlab function generates correct compact PIV data.
