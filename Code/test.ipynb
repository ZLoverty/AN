```python
from pivLib import apply_mask, to_matrix
from skimage import io
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import numpy as np
```

```python
pivData = pd.read_csv("test_files/piv-test.csv")
mask = io.imread("test_files/mask/mask.tif")
p = apply_mask(pivData, mask)
```

```python
p.loc[~p["mask"], ["u", "v"]] = np.nan
```

```python
x, y, u, v = to_matrix(p)
```

```python
plt.imshow(mask, cmap="gray")
plt.quiver(x, y, u, v, color="yellow")
```
