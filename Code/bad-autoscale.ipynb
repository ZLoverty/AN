# Bad autoscale

This notebook illustrates the bad autoscale that cause problems in my PIV flow rate analysis. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = np.mgrid[0:20:50j, 0:20:50j]
normal_image = np.sin(X) - np.cos(Y)
abnormal_image = normal_image.copy()
abnormal_image[10, 10] = 10
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].imshow(normal_image, cmap="gray")
ax[0].set_title("normal image")
ax[1].imshow(abnormal_image, cmap="gray")
ax[1].set_title("abnormal image")
```
